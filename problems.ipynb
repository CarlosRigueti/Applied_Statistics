### Part A: Original Tea Experiment (8 Cups of Tea)
The original Lady Tasting Tea experiment involved 8 cups, 4 tea-first and 4 milk-first. We will calculate the probability of correctly identifying the tea-first cups by chance.



- Part A: Original Tea Experiment (8 Cups of Tea). The original Lady Tasting Tea experiment involved 8 cups, 4 tea-first and 4 milk-first. We will calculate the probability of correctly identifying the tea-first cups by chance.

---


````python
# Number of cups in the original experiment:
no_cups = 8
no_cups_milk_first = 4
no_cups_tea_first = 4
````

````python
# Calculate the number of combinations (order does not matter, no element is selected more than once):
comb = math.comb(no_cups, no_cups_milk_first)
print("The number of possible combinations is:", comb)
````

````python
# Calculate the probability of selecting the correct cups
prob = 1 / comb
print('Probability of correctly identifying the Original Tea Experiment (8 Cups of Tea) is:', prob)
````

````python
# Simulation using NumPy to verify the theoretical probability
# Set up the array representing the cups (1 = tea-first, 0 = milk-first)
cups_8 = np.array([1]*4 + [0]*4)

# Number of simulations
trials = 150000
successes = 0

# Simulation loop
for _ in range(trials):
    np.random.shuffle(cups_8)  # Shuffle the cups randomly
    guessed_cups = cups_8[np.random.choice(range(8), size=4, replace=False)]  # Randomly select 4 cups
    
    if np.sum(guessed_cups) == 4:  # Check if all selected cups are tea-first
        successes += 1

# Estimate probability from the simulation
simulated_probability = successes / trials
print("Estimated probability from simulation:", simulated_probability)
````

## Problem 2: Normal Distribution:

In this problem, we examine how sampling variability affects estimates of the standard deviation. Specifically, we generate 100,000 samples, each consisting of 10 values drawn from a standard normal distribution. For each of these samples, we compute two different types of standard deviations:

1. Sample Standard Deviation (using ddof=1)

2. Population Standard Deviation (using ddof=0)

We then plot the histograms of these two standard deviations on the same axes to visualize their differences and explain the results. We’ll also discuss how the gap between these two estimators changes if the sample size is increased.

### Libraries

For this problem, we will use the following libraries:

NumPy: For generating the samples and performing numerical operations.
- Reference: https://numpy.org/doc/stable/reference/index.html#reference

Matplotlib: For plotting the histograms and visualizing the results.
- Reference: https://matplotlib.org/stable/contents.html


````python
# For numerical operations and generating random samples
import numpy as np  # For numerical operations

# For plotting histograms and visualizations
import matplotlib.pyplot as plt  # For plotting
````

- NumPy provides a function np.random.randn() to generate random samples from a standard normal distribution.

- Matplotlib will allow us to visualize the data by plotting histograms.

----

### Part A: Generating Data
1. Set Parameters:

In this part of the problem, we generate 100,000 samples, each of size 10. The number of samples (num_samples) and the sample size (sample_size) are defined as follows:

````python
# Number of samples we will generate
num_samples = 100000  

# Each sample will have 10 values drawn from the normal distribution
sample_size = 10        
````

- num_samples = 100000: We will generate 100,000 separate samples.

- sample_size = 10: Each sample will contain 10 values drawn from a standard normal distribution.

Thus, our dataset will have the shape (100,000 x 10), meaning 100,000 samples, each consisting of 10 values.

2. Generate Samples from Standard Normal Distribution:

Next, we use NumPy to generate the random samples. The function np.random.randn() generates random numbers from a standard normal distribution, where the mean is 0 and the standard deviation is 1.

````python
samples = np.random.randn(num_samples, sample_size)
````

- samples: This will be a 2D array of size (100,000, 10), where each row corresponds to a sample of size 10.

Each value in samples will be drawn from a standard normal distribution.

3. Compute the Sample Standard Deviation (ddof=1):

Now, we calculate the sample standard deviation using ddof=1. The ddof=1 is used to apply the formula for sample standard deviation, which divides by n-1 (where n is the sample size). This correction is made to ensure that the sample standard deviation is an unbiased estimator of the population standard deviation.

````python
sd_sample = samples.std(axis=1, ddof=1)
````

- axis=1: This tells NumPy to compute the standard deviation across each row (sample).

- ddof=1: This ensures that we calculate the sample standard deviation (correcting for bias by using n-1 in the denominator).

The resulting sd_sample will be an array of 100,000 sample standard deviations.

4. Compute the Population Standard Deviation (ddof=0):

Next, we calculate the population standard deviation using ddof=0. This is the formula for the population standard deviation, which divides by n (the sample size). This formula tends to underestimate the standard deviation for small sample sizes.

````python
sd_population = samples.std(axis=1, ddof=0)
````

- ddof=0: This applies the population standard deviation formula, dividing by n.

The resulting sd_population will be an array of 100,000 population standard deviations.

5. Check the Data Structure:

We can print the first few rows of the generated samples array to visually check the structure and ensure that the data looks correct.

````python
print(samples[:5])  # Prints the first 5 samples to check the structure

# This helps confirm that the data was generated as expected.
````

### Part B: Histogram Comparison

In this part, we visualize the distribution of both the sample and population standard deviations using histograms. The goal is to compare how the two estimators behave, particularly when the sample size is small.

1. Plot the Histograms:

We will use Matplotlib to create histograms for both the sample standard deviation and population standard deviation. The histograms will be plotted on the same axes with transparency (alpha=0.6) to make it easier to compare the two distributions.

````python
# Create a figure with size 10x6
plt.figure(figsize=(10,6))  
plt.hist(sd_sample, bins=60, alpha=0.6, label="Sample SD (ddof=1)", color='blue')
plt.hist(sd_population, bins=60, alpha=0.6, label="Population SD (ddof=0)", color='red')

# Title of the plot
plt.title("Distribution of Sample SD vs. Population SD (n=10, 100,000 samples)")  
# X-axis label
plt.xlabel("Standard Deviation") 
# X-axis label
plt.ylabel("Frequency") 
# Add a legend to differentiate the two histograms
plt.legend() 
# Add grid with low transparency for better visibility
plt.grid(alpha=0.3) 
# Display the plot
plt.show()  
````
