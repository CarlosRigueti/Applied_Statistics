The original Lady Tasting Tea experiment involved 8 cups, 4 tea-first and 4 milk-first. We will calculate the probability of correctly identifying the tea-first cups by chance.



**- Part A**: Original Tea Experiment (8 Cups of Tea). The original Lady Tasting Tea experiment involved 8 cups, 4 tea-first and 4 milk-first. We will calculate the probability of correctly identifying the tea-first cups by chance.

---


````python
# Number of cups in the original experiment:
no_cups = 8
no_cups_milk_first = 4
no_cups_tea_first = 4
````

````python
# Calculate the number of combinations (order does not matter, no element is selected more than once):
comb = math.comb(no_cups, no_cups_milk_first)
print("The number of possible combinations is:", comb)
````

````python
# Calculate the probability of selecting the correct cups
prob = 1 / comb
print('Probability of correctly identifying the Original Tea Experiment (8 Cups of Tea) is:', prob)
````

````python
# Simulation using NumPy to verify the theoretical probability
# Set up the array representing the cups (1 = tea-first, 0 = milk-first)
cups_8 = np.array([1]*4 + [0]*4)

# Number of simulations
trials = 150000
successes = 0

# Simulation loop
for _ in range(trials):
    np.random.shuffle(cups_8)  # Shuffle the cups randomly
    guessed_cups = cups_8[np.random.choice(range(8), size=4, replace=False)]  # Randomly select 4 cups
    
    if np.sum(guessed_cups) == 4:  # Check if all selected cups are tea-first
        successes += 1

# Estimate probability from the simulation
simulated_probability = successes / trials
print("Estimated probability from simulation:", simulated_probability)
````

### **Problem 2**: Normal Distribution:

In this problem, we examine how sampling variability affects estimates of the standard deviation. Specifically, we generate 100,000 samples, each consisting of 10 values drawn from a standard normal distribution. For each of these samples, we compute two different types of standard deviations:

1. Sample Standard Deviation (using ddof=1)

2. Population Standard Deviation (using ddof=0)

We then plot the histograms of these two standard deviations on the same axes to visualize their differences and explain the results. We’ll also discuss how the gap between these two estimators changes if the sample size is increased.

---

For this problem, we will use the following libraries:

````python
# For numerical operations and generating random samples
import numpy as np  # For numerical operations

# For plotting histograms and visualizations
import matplotlib.pyplot as plt  # For plotting
````

- NumPy provides a function np.random.randn() to generate random samples from a standard normal distribution.

- Matplotlib will allow us to visualize the data by plotting histograms.

### **Part A**: Generating Data
1. Set Parameters:

In this part of the problem, we generate 100,000 samples, each of size 10. The number of samples (num_samples) and the sample size (sample_size) are defined as follows:

````python
# Number of samples we will generate
num_samples = 100000  

# Each sample will have 10 values drawn from the normal distribution
sample_size = 10        
````

- _num_samples_ = 100000: We will generate 100,000 separate samples.

- _sample_size_ = 10: Each sample will contain 10 values drawn from a standard normal distribution.

Thus, our dataset will have the shape (100,000 x 10), meaning 100,000 samples, each consisting of 10 values.

---

2. Generate Samples from Standard Normal Distribution:

Next, we use NumPy to generate the random samples. The function _np.random.randn()_ generates random numbers from a standard normal distribution, where the mean is 0 and the standard deviation is 1.

````python
samples = np.random.randn(num_samples, sample_size)
````

- _samples_: This will be a 2D array of size (100,000, 10), where each row corresponds to a sample of size 10.

Each value in samples will be drawn from a standard normal distribution.

---

3. Compute the Sample Standard Deviation (ddof=1):

Now, we calculate the sample standard deviation using ddof=1. The ddof=1 is used to apply the formula for sample standard deviation, which divides by n-1 (where n is the sample size). This correction is made to ensure that the sample standard deviation is an unbiased estimator of the population standard deviation.

````python
sd_sample = samples.std(axis=1, ddof=1)
````

- **_axis=1_**: This tells NumPy to compute the standard deviation across each row (sample).

- **_ddof=1_**: This ensures that we calculate the sample standard deviation (correcting for bias by using n-1 in the denominator).

The resulting sd_sample will be an array of 100,000 sample standard deviations.

----

4. Compute the Population Standard Deviation **(ddof=0)**:

Next, we calculate the population standard deviation using ddof=0. This is the formula for the population standard deviation, which divides by n (the sample size). This formula tends to underestimate the standard deviation for small sample sizes.

````python
sd_population = samples.std(axis=1, ddof=0)
````

- **_ddof=0_**: This applies the population standard deviation formula, dividing by _n_.

The resulting sd_population will be an array of 100,000 population standard deviations.

---

5. Check the Data Structure:

We can print the first few rows of the generated samples array to visually check the structure and ensure that the data looks correct.

````python
print(samples[:5])  # Prints the first 5 samples to check the structure

# This helps confirm that the data was generated as expected.
````

### **Part B**: Histogram Comparison

In this part, we visualize the distribution of both the sample and population standard deviations using histograms. The goal is to compare how the two estimators behave, particularly when the sample size is small.

#### 1. **Plot the Histograms**:

We will use Matplotlib to create histograms for both the sample standard deviation and population standard deviation. The histograms will be plotted on the same axes with transparency (alpha=0.6) to make it easier to compare the two distributions.

````python
# Create a figure with size 10x6
plt.figure(figsize=(10,6))  
plt.hist(sd_sample, bins=60, alpha=0.6, label="Sample SD (ddof=1)", color='blue')
plt.hist(sd_population, bins=60, alpha=0.6, label="Population SD (ddof=0)", color='red')

# Title of the plot
plt.title("Distribution of Sample SD vs. Population SD (n=10, 100,000 samples)")  
# X-axis label
plt.xlabel("Standard Deviation") 
# X-axis label
plt.ylabel("Frequency") 
# Add a legend to differentiate the two histograms
plt.legend() 
# Add grid with low transparency for better visibility
plt.grid(alpha=0.3) 
# Display the plot
plt.show()  
````


**- _Explanation_**: The **histograms** of the sample SD and population SD allow us to visualize the inherent differences between the two estimators. The **sample SD** is slightly larger on average due to the bias correction _(ddof=1)_, and it exhibits more spread, especially for smaller sample sizes. As sample sizes increase, the difference between these two distributions decreases, and they converge to the population standard deviation.

---

#### 2. **Analysis**:

From the histogram, we can analyze the differences between the sample SD and population SD:

Sample Standard Deviation (using ddof=1) will generally be slightly larger than the population SD (using ddof=0), especially for small samples. This happens because the sample SD compensates for the bias in estimating the population standard deviation.

Population Standard Deviation (using ddof=0) tends to underestimate the true standard deviation, particularly for small sample sizes.

#### 3. **Mean Comparison**:

We can also compare the mean values of both standard deviations:

````python
print("Sample SD (ddof=1) mean:", np.mean(sd_sample))

print("Population SD (ddof=0) mean:", np.mean(sd_population))
````

- The mean of the sample SD will be slightly higher than the mean of the population SD due to the correction for bias.

---

### **Conclusion**:

From the histogram and the statistical comparison, we observe the following:

1. Sample SD **(ddof=1)** tends to be slightly larger than the Population SD **(ddof=0)**, particularly when the sample size is small. This is because the sample standard deviation is an unbiased estimator, whereas the population standard deviation tends to underestimate the true variability.

2. As the sample size increases, the difference between the sample SD and the population SD shrinks, and both estimators become nearly identical for large samples.

3. The histograms show a clear visual difference in the two distributions, with the sample SD distribution being slightly shifted to the right **(larger values)**.

This analysis demonstrates the impact of sample size on the estimation of variability and helps us understand the behavior of different standard deviation estimators.

### **Problem 3**: t-Tests and Type II Errors

- In this problem, we are tasked with investigating Type II errors in hypothesis testing. A Type II error occurs when a statistical test fails to reject the null hypothesis even though it is actually false. In other words, it’s a **“miss”** — the test didn’t detect a real difference or effect when one actually exists.

- We will simulate the independent samples t-test and examine how the Type II error rate changes as the difference in means between two samples increases.

### **Problem Breakdown**:

- We are asked to simulate the t-test for different mean differences d (ranging from 0 to 1 in steps of 0.1).

- For each value of d, we will:
    1. Draw two samples of size 100:

      - One sample is from the standard normal distribution (mean = 0, standard deviation = 1).

      - The other sample is from the normal distribution with mean d and standard deviation 1.

    2. Run an independent samples t-test to compare the two samples.

    3. Reject the null hypothesis if the p-value is less than 0.05.

    4. Record the proportion of times the null hypothesis is not rejected (which indicates a Type II error).

    5. Plot the proportion of Type II errors against d, and explain how the Type II error rate changes as the difference in means increases

#### **Steps**:

1. Simulate the Data: For each mean difference d, we generate two sets of samples, each with 100 data points.

2. Conduct t-Tests: Perform an independent samples t-test to compare the two samples.

3. Calculate Proportion of Type II Errors: For each mean difference, calculate the proportion of times the null hypothesis is not rejected (i.e., the Type II error rate).

4. Plot the Results: Plot the proportion of Type II errors against d, and analyze the trend.

#### **Libraries Needed**:

We will use the following Python libraries:

 - NumPy: For generating random samples and numerical operations.

 - SciPy: For conducting the t-test.

 - Matplotlib: For plotting the results.


### **Code Implementation**:

````python

from scipy import stats  # For performing t-test


# Parameters
num_simulations = 1000  # Number of simulations to run for each d
sample_size = 100  # Size of each sample
alpha = 0.05  # Significance level for the t-test

# Array to store the proportion of Type II errors for each mean difference d
type_ii_errors = []

# Range of mean differences d (from 0 to 1, step 0.1)
mean_differences = np.arange(0, 1.1, 0.1)

# Simulating for each mean difference d
for d in mean_differences:
    errors = 0  # Counter for Type II errors
    
    # Run the simulation 1000 times
    for _ in range(num_simulations):
        # Generate two samples:
        # 1. Sample from standard normal (mean = 0, std = 1)
        sample1 = np.random.normal(loc=0, scale=1, size=sample_size)
        
        # 2. Sample from normal distribution with mean = d and std = 1
        sample2 = np.random.normal(loc=d, scale=1, size=sample_size)
        
        # Perform the independent samples t-test
        t_stat, p_value = stats.ttest_ind(sample1, sample2)
        
        # If p-value is greater than 0.05, fail to reject the null hypothesis (Type II error)
        if p_value > alpha:
            errors += 1
    
    # Calculate the proportion of Type II errors
    type_ii_error_rate = errors / num_simulations
    type_ii_errors.append(type_ii_error_rate)

# Plotting the results
plt.figure(figsize=(10,6))
plt.plot(mean_differences, type_ii_errors, marker='o', linestyle='-', color='b')
plt.title('Proportion of Type II Errors vs. Mean Difference (d)')
plt.xlabel('Mean Difference (d)')
plt.ylabel('Proportion of Type II Errors')
plt.grid(True)
plt.show()
````

---

#### **Code Explanation**:

1. **Setting Parameters**:

   - _num_simulations = 1000_: We repeat the simulation 1,000 times for each value of d.

   - _sample_size_ = 100: Each sample consists of 100 data points.

   - _alpha = 0.05_: The significance level (p-value threshold) for the t-test is _0.05_, which is commonly used in hypothesis testing.

2. **Simulating the Data**:

   - For each mean difference d in the range from 0 to 1 _(in steps of 0.1)_, we generate two samples:

   - The first sample is drawn from the standard normal distribution with _mean = 0_ and standard deviation _= 1_.

   - The second sample is drawn from the normal distribution with _mean = d_ and standard deviation _= 1_.

3. **Conducting the t-Test**:

   - We use stats.ttest_ind() from SciPy to perform an independent samples _t-test_ on the two samples. This tests whether the means of the two samples are significantly different.

   - If the p-value from the t-test is greater than _0.05_, we fail to reject the null hypothesis, which is considered a _Type II_ error

4. **Tracking Type II Errors**:

   - For each simulation, if we fail to reject the null hypothesis, we increment the error count.

   - After running 1,000 simulations for a given d, we compute the proportion of _Type II_ errors _(i.e., the ratio of times the null hypothesis was not rejected)_.

5. **Plotting the Results**:

   - Finally, we plot the proportion of _Type II_ errors against the mean difference _d._ This shows how the error rate changes as the true mean difference between the two samples increases.

 ---


#### **Expected Results and Interpretation**:

   - **When d=0** _(i.e., there is no true difference between the two samples)_, the proportion of Type II errors will be high because the null hypothesis (that there is no difference) is true, but the t-test might still fail to reject it due to random variability.

   - As _d_ increases, the true difference between the two samples becomes larger, making it easier to detect the difference. Therefore, the proportion of Type II errors should decrease as d increases, meaning that the test is more likely to reject the null hypothesis when there is a larger true difference.

 ---

#### **Conclusion**:

   - This simulation shows how the Type II error rate changes as the true difference between the means increases. When the difference is small, the error rate is high because it’s harder for the t-test to detect a difference. As the difference grows, the error rate decreases, and the test becomes more sensitive to differences.

   - Understanding Type II errors is crucial for determining the power of a statistical test. A low Type II error rate indicates that the test is sensitive and capable of detecting real differences when they exist.

### Problem 4: ANOVA (Analysis of Variance):

In this problem, we are tasked with performing a one-way ANOVA to test whether the means of three independent samples drawn from normal distributions are equal. Additionally, we will perform three independent two-sample t-tests to compare the means of the pairs of samples and compare the conclusions. Finally, we will discuss why ANOVA is preferred over conducting multiple t-tests.

#### Understanding the Problem:

**ANOVA** (Analysis of Variance) is a statistical method used to compare the means of more than two groups. In this case, we will compare the means of three samples. The null hypothesis (H₀) in **ANOVA** is that all the group means are equal, and the alternative hypothesis (H₁) is that at least one group mean is different.

On the other hand, t-tests are typically used to compare the means of two groups at a time. We are asked to perform three t-tests:

1. Sample 1 vs. Sample 2

2. Sample 1 vs. Sample 3

3. Sample 2 vs. Sample 3

After performing these t-tests, we will compare the results with the **ANOVA** conclusion and discuss the advantages of using **ANOVA**.

#### Steps to Solve the Problem:

1. Generate the Data: We will generate three independent samples from normal distributions with means of 0, 0.5, and 1, and a standard deviation of 1. Each sample will contain 30 values.

2. Perform One-Way **ANOVA**: We will use the **ANOVA** test to compare the means of the three samples and check if they are significantly different.

3. Perform Two-Sample t-Tests: We will conduct three independent two-sample t-tests to compare the means of the sample pairs.

4. Compare the Results: We will compare the conclusions from the **ANOVA** and the t-tests and explain why **ANOVA** is generally preferred over multiple t-tests.

#### Step-by-Step Solution:

1. Generate the Data

First, let's generate the three independent samples. Each sample will be of size 30, drawn from normal distributions with means 0, 0.5, and 1, and standard deviation 1.

````python
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate three independent samples
sample_size = 30
mean_1, mean_2, mean_3 = 0, 0.5, 1
std_dev = 1

# Generate samples from normal distributions with the specified means and standard deviations
sample_1 = np.random.normal(loc=mean_1, scale=std_dev, size=sample_size)
sample_2 = np.random.normal(loc=mean_2, scale=std_dev, size=sample_size)
sample_3 = np.random.normal(loc=mean_3, scale=std_dev, size=sample_size)

# Display the first few entries of each sample
print("Sample 1 (mean = 0):", sample_1[:5])
print("Sample 2 (mean = 0.5):", sample_2[:5])
print("Sample 3 (mean = 1):", sample_3[:5])
````

**_- Explanation_**: Here, we generate three samples, each containing 30 data points. The samples are drawn from normal distributions with means of 0, 0.5, and 1, respectively, and a standard deviation of 1.

2. Perform One-Way ANOVA

Now, let's perform a one-way ANOVA to test if the means of the three samples are significantly different.

````python
from scipy import stats

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(sample_1, sample_2, sample_3)

# Display the results
print(f"ANOVA F-statistic: {f_statistic}")
print(f"ANOVA p-value: {p_value}")
````

**_- Explanation_**: We use scipy.stats.f_oneway() to perform the **one-way ANOVA** test. The function returns the F-statistic and the p-value. The **F-statistic** measures the ratio of variances between the groups, and the **p-value** tells us whether the differences in means are statistically significant.

3. Perform Two-Sample t-Tests

Next, let's perform three independent two-sample t-tests to compare the means of the pairs of samples.

````python
# Perform two-sample t-tests
t_stat_1_2, p_value_1_2 = stats.ttest_ind(sample_1, sample_2)
t_stat_1_3, p_value_1_3 = stats.ttest_ind(sample_1, sample_3)
t_stat_2_3, p_value_2_3 = stats.ttest_ind(sample_2, sample_3)

# Display the results
print(f"T-test 1 vs 2: t-statistic = {t_stat_1_2}, p-value = {p_value_1_2}")
print(f"T-test 1 vs 3: t-statistic = {t_stat_1_3}, p-value = {p_value_1_3}")
print(f"T-test 2 vs 3: t-statistic = {t_stat_2_3}, p-value = {p_value_2_3}")
````

**-_Explanation_**: We perform three independent **two-sample t-tests** using scipy.stats.ttest_ind(). This function returns the **t-statistic** and **p-value** for each pair of samples:

1. Sample 1 vs. Sample 2

2. Sample 1 vs. Sample 3

3. Sample 2 vs. Sample 3

4. Compare the Results

We compare the **p-values** from the **ANOVA** test and the t-tests:

- If the **p-value** is less than **0.05**, we reject the null hypothesis, indicating a significant difference between the means.

- If the p-value is greater than **0.05**, we fail to reject the null hypothesis, suggesting that there is no significant difference.

**Based on the results**:

**- ANOVA** will provide a single p-value that tests whether all group means are equal. If the p-value is below **0.05**, we reject the null hypothesis, concluding that at least one group mean differs from the others.

**- The t-tests** test pairwise comparisons between the groups. If any p-value is less than **0.05**, we conclude that the means of those two groups are significantly different.

5. Plot the Results

- Finally, we can visualize the results of the **ANOVA** test and the t-tests. For simplicity, let’s create a bar plot of the means of the three samples to give us a sense of the differences.

````python
# Plot the means of the three samples
means = [np.mean(sample_1), np.mean(sample_2), np.mean(sample_3)]
labels = ['Sample 1', 'Sample 2', 'Sample 3']

plt.bar(labels, means, color=['blue', 'orange', 'green'])
plt.title('Means of the Three Samples')
plt.xlabel('Sample')
plt.ylabel('Mean')
plt.show()
````

### **Conclusion**: Why ANOVA is Preferred Over t-Tests

**- ANOVA** vs. t-tests: While t-tests can be used to compare the means of two samples, when you have three or more samples, performing multiple t-tests increases the risk of Type I errors (false positives). This happens because each t-test has its own 5% chance of incorrectly rejecting the null hypothesis, and when performed multiple times, the overall probability of incorrectly rejecting the null hypothesis increases.

**- ANOVA** is designed to test all the groups at once, reducing the overall Type I error rate. By using a single statistical test, ANOVA controls the error rate more effectively and provides a clearer picture of whether there is any significant difference among the groups.

In summary, ***ANOVA** is preferred when comparing the means of three or more groups because it allows for a more robust and reliable test without inflating the error rate, as multiple t-tests would.

**Addition**: Discuss the **family-wise error rate (FWER)** when performing multiple t-tests and why ANOVA is a better approach. Multiple t-tests increase the risk of **false positives** (rejecting a true null hypothesis), and this is known as the **family-wise error rate**. By using ANOVA, you control the overall error rate, avoiding the inflation of false positives that comes with multiple comparisons.

### Family-Wise Error Rate and Why ANOVA is Preferred:
When performing multiple t-tests, the **family-wise error rate (FWER)** increases, meaning the probability of committing at least one Type I error (false positive) across multiple tests becomes larger. This is because each t-test carries a 5% chance of falsely rejecting the null hypothesis, and when performed multiple times, the overall error rate grows. **ANOVA** controls for this by testing all groups at once, reducing the risk of Type I errors. This makes ANOVA a more reliable method for comparing three or more groups.
